```python
import pandas as pd
```

```python
# Reading in the items to replace:
replacing_dict = {}
replacing_file = open('./full_opinion_list.txt', 'r')
for each_line in replacing_file:
    info_list = each_line.strip().split('\t')
    replacing_dict[info_list[0]] = info_list[2]
```

```python
# Reading in the files:
years = ['1999', '2001', '2003', '2005', '2007',
         '2009', '2011', '2013', '2015', '2017', 
         '2019', '2000', '2002', '2004', '2006',
         '2008', '2010', '2012', '2014', '2016', 
         '2018']

for eachYear in years: 
    temp_file = pd.read_excel('./parsed_wiki_data/%s.xlsx' % str(eachYear), index_col=0)
    temp_file.index = list(range(temp_file.shape[0]))
    temp_file.replace(replacing_dict, inplace=True)
    temp_file.to_excel('./parsed_wiki_data_simpler/%s.xlsx' % str(eachYear))
```
